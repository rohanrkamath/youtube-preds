# Cleaning

```python
def preprocess_inputs(df):
    df = df.copy()
    
    # dropping useless columns
    df = df.drop(['video_id', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'], axis=1)
    
    # remove duplicates
    df = df.drop_duplicates()
    
    # remove rows with null values
    df = df.dropna(how='any',axis=0) 
    
#     # convert 'publish_time' to get months (I feel which month of posting might make a difference)
#     df['publish_time'] = pd.to_datetime(df['publish_time'])
#     df['month_of publishing'] = df['publish_time'].apply(lambda x: x.month)
#     df = df.drop(['publish_time', 'trending_date'], axis=1)
    
#     # get tag count
#     tag_count = list()
#     for i in df['tags']:
#         j = i.split('|')
#         count = len(j)
#         tag_count.append(count)
        
#     df['tag_count'] = tag_count
    
    return df
```

```python
data = preprocess_inputs(raw_data)
```

# Features

```python
# convert 'publish_time' to get months (I feel which month of posting might make a difference)

# data['publish_time'] = pd.to_datetime(data['publish_time'])
# data['month_of publishing'] = data['publish_time'].apply(lambda x: x.month)
# data = data.drop(['publish_time', 'trending_date'], axis=1)

# data

publish_time= pd.to_datetime(data['publish_time'])
data['publish_time_1'] = publish_time.dt.time
data['publish_date'] = publish_time.dt.date

# day at which video is publish
data['publish_weekday']=publish_time.dt.strftime('%A')
data
```

```python
# length of title
title_len=[len(x) for x in data["title"]]
data["title_length"]=title_len
data
```

```python
#length of description
desc_len=[len(x) for x in data["description"]]
data["description_length"]=desc_len
data
```

```python
# Ratios 
# ratio of view/likes
data["Ratio_View_likes"]=round(data["views"]/data["likes"],3)
# ratio of view/dislikes 
data["Ratio_View_dislikes"]=round(data["views"]/data["dislikes"],3)
# ratio of view/comment_count
data["Ratio_views_comment_count"]=round(data["views"]/data["comment_count"],3)
# ratio of likes/dislikes
data["Ratio_likes_dislikes"]=round(data["likes"]/data["dislikes"],3)
```

```python
# remove infinite values
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna(how='any',axis=0)

data
```

```python
data.drop(['trending_date','publish_date','publish_time','tags','title','description','channel_title'],axis=1,inplace=True)
```
